```python
# check class imbalance
train_folder_path = image_path / "Train"

subdirectories = [f.path for f in os.scandir(train_folder_path) if f.is_dir() ]

train_image_dict = {}

for subdir in subdirectories:
    class_name = os.path.split(subdir)[1]
    
    train_image_dict[class_name] = os.listdir(subdir)

for key, value in train_image_dict.items():
    print(f"Class : {key} | Total images : {len(value)}")
```
